```python
# Lib Imports
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate weighted graph of road network
road_map = nx.Graph()
t = 10
elist = [('start', 'dest', t)]
road_map.add_weighted_edges_from(elist)


nx.draw(road_map)
```
